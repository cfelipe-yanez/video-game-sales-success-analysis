```python
# Convierte los datos en los tipos necesarios.
df['year_of_release'] = pd.to_numeric(df['year_of_release'], errors='coerce').astype('Int64')
if (df['critic_score'] % 1 == 0).all() == 0:
    df['critic_score'] = pd.to_numeric(df['critic_score'], errors='coerce').astype('Int64')
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce').astype('float64')
df.info()
df.sample(20)
```

Hasta ahora he cambiado los nombres de las columnas que ahora todos están en miniscula e hice tres cambios combertí dos de las columnas en Int64 pues no tenían decimales entonces para que no queden .00 y el user_score estaba inicialmente en 

```python
#tratar los valores ausentes
if (df['critic_score'].isna().sum() / len(df) * 100 > 20):
    print(f"El porcentaje de critic_score con valor nulo es:{df['critic_score'].isna().sum() / len(df) * 100:.2f}%, etonces se rellena con la media para no perder tantos datos")
    df['critic_score'] = df['critic_score'].fillna(round(df['critic_score'].mean()))

if (df['user_score'].isna().sum() / len(df) * 100  > 20):
    print(f"El porcentaje de user_score con valor nulo es: {df['user_score'].isna().sum() / len(df) * 100:.2f}%, etonces se rellena con la media para no perder tantos datos")
    df['user_score'] = df['user_score'].fillna(round(df['user_score'].mean()))    

    
    
if (df['rating'].isna().sum() / len(df) * 100  > 20):
    print(f"El porcentaje de user_score con valor nulo es: {df['rating'].isna().sum() / len(df) * 100:.2f}%, etonces se rellena con la media para no perder tantos datos")
    df['rating'] = df['rating'].fillna(df['rating'].mode().iloc[0])   

df.sample(20)
```

Hipotesis nula: Las calificaciones promedio de los usuarios para Xbox One y PC son iguales.

Hipotesis Alternativa:  Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.

Alfa: 0.05

```python
xbox_one_user_scores = df[df['platform'] == 'XOne']['user_score'].dropna()
pc_user_scores = df[df['platform'] == 'PC']['user_score'].dropna()
alpha = 0.05

# Prueba t de Student
t_stat_1, p_value_1 = st.ttest_ind(xbox_one_user_scores, pc_user_scores)

hypothesis_1_result = "Rechazar H₀" if p_value_1 < alpha else "No rechazar H₀"
t_stat_1, p_value_1, hypothesis_1_result
```

## Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes

Hipotesis nula: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son iguales.

Hipotesis Alternativa:   Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.


Alfa: 0.05

```python
accion_user_scores = df[df['genre'] == 'Action']['user_score'].dropna()
deportes_user_scores = df[df['genre'] == 'Sports']['user_score'].dropna()

# Prueba t de Student
t_stat_2, p_value_2 = st.ttest_ind(accion_user_scores, deportes_user_scores)
hypothesis_2_result = "Rechazar H₀" if p_value_2 < alpha else "No rechazar H₀"

t_stat_2, p_value_2, hypothesis_2_result
```

```python
# 1. Filtrar y convertir las calificaciones a numéricas
accion_user_scores = df[df['genre'] == 'Action']['user_score']
deportes_user_scores = df[df['genre'] == 'Sports']['user_score']

# Asegúrate de que estén en formato numérico (en caso de que haya "tbd" u otros strings)
accion_user_scores = pd.to_numeric(accion_user_scores, errors='coerce').dropna()
deportes_user_scores = pd.to_numeric(deportes_user_scores, errors='coerce').dropna()

# 2. Prueba de Levene para comparar varianzas
stat_levene, p_levene = st.levene(accion_user_scores, deportes_user_scores)

# Determinar si las varianzas son iguales
equal_var = p_levene > 0.05  # Si p > 0.05, asumimos varianzas iguales

# 3. Prueba t de Student (con ajuste según resultado de Levene)
t_stat_2, p_value_2 = st.ttest_ind(
    accion_user_scores,
    deportes_user_scores,
    equal_var=equal_var
)

# 4. Interpretación
alpha = 0.05
hypothesis_2_result = "Rechazar H₀" if p_value_2 < alpha else "No rechazar H₀"

# Resultados
print(f"Prueba de Levene: p = {p_levene:.5f} -> {'Varianzas iguales' if equal_var else 'Varianzas diferentes'}")
print(f"T-test: t = {t_stat_2:.4f}, p = {p_value_2:.5f} -> {hypothesis_2_result}")
```

# Verificar los supuestos de las pruebas de hipótesis.

```python
# 1. Visualización de distribuciones
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(accion_user_scores, kde=True, color='blue', label='Action')
sns.histplot(deportes_user_scores, kde=True, color='orange', label='Sports')
plt.title('Distribución de user_score')
plt.legend()

plt.subplot(1, 2, 2)
sns.boxplot(data=[accion_user_scores, deportes_user_scores], palette='Set2')
plt.xticks([0, 1], ['Action', 'Sports'])
plt.title('Boxplot de user_score')

plt.tight_layout()
plt.show()

# 2. Prueba de normalidad (Shapiro-Wilk)
shapiro_action = st.shapiro(accion_user_scores.sample(500, random_state=1))
shapiro_sports = st.shapiro(deportes_user_scores.sample(500, random_state=1))

print(f"Shapiro Action: p = {shapiro_action.pvalue:.5f}")
print(f"Shapiro Sports: p = {shapiro_sports.pvalue:.5f}")

# 3. Prueba de igualdad de varianzas (Levene)
levene_stat, levene_p = st.levene(accion_user_scores, deportes_user_scores)
print(f"Levene test: p = {levene_p:.5f}")

# 4. Prueba t de Student
equal_var = True if levene_p > 0.05 else False
t_stat, p_value = st.ttest_ind(accion_user_scores, deportes_user_scores, equal_var=equal_var)

alpha = 0.05
resultado = "Rechazar H₀" if p_value < alpha else "No rechazar H₀"

print(f"\nPrueba t:")
print(f"t = {t_stat:.4f}, p = {p_value:.5f}")
print(f"Resultado: {resultado}")
```
